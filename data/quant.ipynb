```python
import pandas as pd

# Define the table structure
df = pd.DataFrame(columns=['name', 'age', 'gender'])

# An array of dictionaries to iterate over
data = [
    {'name': 'Alice', 'age': 25, 'gender': 'Female'},
    {'name': 'Bob', 'age': 30, 'gender': 'Male'},
    {'name': 'Charlie', 'age': 35, 'gender': 'Male'},
]

# Iterate over the array of dictionaries and append each key and value to the table
for d in data:
    df = df.append({'name': d['name'], 'age': d['age'], 'gender': d['gender']}, ignore_index=True)

# Print the resulting table
print(df)
```
